```python
import torch
import numpy as np
import torch.nn as nn
```

```python
pretrained_mobilenet_model = torch.hub.load('pytorch/vision:v0.6.0', 'mobilenet_v2', pretrained=True)
```

```python
class Net(nn.Module):
    def __init__(self, my_pretrained_model):
        super(Net, self).__init__()
        self.pretrained = my_pretrained_model
        for name,params in self.pretrained.named_parameters():
            params.requires_grad = False
        self.my_new_layers = nn.Sequential(nn.(1000, 100),
                                           nn.ReLU(),
                                           nn.Linear(100, 2))
    
    def forward(self, x):
        x = self.pretrained(x)
        x = self.my_new_layers(x)
        return x

reduced_pretrained_network = nn.Sequential(*list(pretrained_mobilenet_model.features.children())[:-1])
my_extended_model = Net(my_pretrained_model=reduced_pretrained_network)
```
